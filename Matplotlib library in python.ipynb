```python
x1 = np.arange(1,11)
x1
```

```python
y1 = 99/x1
y2= x1*3
y1
y2
```

```python
plt.plot(x1,y1,color="r",linestyle =":",linewidth =6)
plt.plot(x1,y2,color ="g",linestyle ="-.",linewidth =6)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x1,y1)
plt.grid()
plt.subplot(1,2,2)
plt.plot(x1,y2)
plt.grid()
plt.show()
```

```python
#piechart
```

```python
items='laptop','mobile','charger','mouse','keyboard'
no_of_items=[5,10,7,5,3]
explode=(0,.2,0,0,0.1)
```

```python
plt.pie(no_of_items,labels=items)
plt.show()
```

```python

plt.pie(no_of_items,labels=items,colors=['black','red','g','b','yellow'],autopct='%0.1f%%',explode=explode )
plt.show()
```

```python
plt.pie(no_of_items,labels=items,radius=2.5)
plt.pie([1],colors='white',radius =1.4)
plt.show()
```
